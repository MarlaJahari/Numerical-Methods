```python
# -*- coding: utf-8 -*-
"""
Solving definite integrals using Simpsons Method

"""
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 1/(1+x**2)

a=float(input("Enter the initial limit: "))
b=float(input("Enter the final limit: "))


#int[f(x)dx]=(h/3)[yo+yn+2*y(even)+4*y(odd)]


n1=int(input("Enter n for (1/3)rd rule:"))
h1=(b-a)/n1
sol1=f(a)+f(b)

for i in range(1,n1):
    if i%2!=0:
        sol1=sol1+4*f(a+i*h1)
    if i%2==0:
        sol1=sol1+2*f(a+i*h1)
        
sim1_3=h1*sol1/3
print("Solution by simpson's (1/3) rule",sim1_3)


n2=int(input("Enter n for (3/8)rd rule:"))
h2=(b-a)/n2
sol1=f(a)+f(b)

for i in range(1,n2):
    if i%3!=0:
        sol1=sol1+3*f(a+i*h2)
    if i%3==0:
        sol1=sol1+2*f(a+i*h2)
        
sim3_8=(3/8)*h2*sol1
print("Solution by simpson's (3/8) rule",sim3_8)


n3=int(input("Enter n (trapezoidal rule): "))
h3=(b-a)/n3
sol1=f(a)+f(b)

sol=0
for i in range(1,n3):
        sol=sol+f(a+i*h3)
        
trap=h3*sol1/2+h3*sol
print("Solution by trapezoidal rule: ",trap)

x=np.linspace(a,b,1000)
plt.plot(x,f(x))
```
