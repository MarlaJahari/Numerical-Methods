```python
"""
PDE by Bender-Schimdt Method

Q- del(u)/del(t)=(del^2)u/(del(x)^2)
    u(0,t)=u(1,t)=0
    u(x,0)=sin(pi*x), x=[0,1]
    h=0.2, k=0.02, t=[0,0.06]
"""
import matplotlib.pyplot as plt
import numpy as np
import math

print("Enter the limits for x")
xi=float(input(""))
xf=float(input(""))
print("Enter the limits for t")
ti=float(input(""))
tf=float(input(""))

#OUR PROBLEM
#uxx(i,j)=u(i+1,j)-2u(i,j)+u(i-1,j)/h^2
#ut(i,j)=u(i,j+1)-u(i,j)/k
#where h=t(i+1)-t(i)
#k=x(i+1)-x(i)
        
h=float(input("enter h\t"))
k=float(input("enter k\t"))
n=int((xf-xi)/h)

#    u(i,j+1)=a*u(i-1,j)+(1-2*a)*u(i,j)+a*u(i+1,j)
#    a=k/h^2 
a=k/h**2                    
x=np.arange(xi,xf+h,h)      #Possible values of x
t=np.arange(ti,tf+k,k)      #Possible values of t

nr,nc=x.size,t.size
U=np.zeros([nr,nc])          #The necessary matrix

for j in range(nc):
    for i in range(1,nr-1):
           if j==0:
                U[i,j]=math.sin(np.pi*x[i])     #Condition for u(x,t)
           else:
                U[i,j]=a*U[i-1,j-1]+(1-2*a)*U[i,j-1]+a*U[i+1,j-1]

                
print("The matrix U containing all the values of u:\n",U)

T=float(input("Value of t at\t"))
X=float(input("Value of x at\t"))

n2=int(T/k)
n1=int(X/h)

print("Value of u(t=",t,",x=",x,")is\n",U[n1,n2])

T,X=np.meshgrid(t,x)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(T,X,U,color='red')
ax.set_xlabel("T")
ax.set_ylabel("X")
ax.set_zlabel("u")
ax.set_title("Solution of heat equation for the asked equation")
plt.savefig("Heat equation.jpg",dpi=1000)
```
