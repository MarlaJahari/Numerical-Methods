```python
import matplotlib.pyplot as plt
import numpy as np

def u(t):
    return 200*t

a=1/16
b=0
c=0
print("Enter the limits for x:")
xi=float(input(""))
xf=float(input(""))

"""
Using the Crank-Nicolson method:

Ai*u(i+1,j+1)+Bi*u(i,j+1)+Ci*u(i-1,j+1)=Di

where Ai=2ka+khb
      Bi=4h^2+4ka-2kch^2
      Ci=khb-2ka
      Di=(2ka+khb)*u(i+1,j)+(4h^2-4ka+2kch^2)*u(i,j)+(2ka-khb)*u(i-1,j)
      h=stepsize for x coordinates
      k=stepsize for t coordinates
for i=[1,n]

"""
h=float(input('Enter h:'))
n1=int((xf-xi)/h)
k=h*h/a             
n2=1        
print(k)
C=-(2*k*a+k*h*b)
B=4*(h*h)+4*k*a-2*k*c*(h*h)
A=k*h*b-2*k*a
X=[]
Y=[]
Z=[]    
for i in range(n1+1):
   Y.append(B)    
for i in range(n1):
   X.append(A)
   Z.append(C)  
M=np.diag(X,-1)+np.diag(Y,0)+np.diag(Z,1)
print("Tri-diag formed\n",M)

x=np.arange(xi,xf+h,h)     
t=np.array([0,k])          #possible values of x & t
print('t',t)

""Since, no.of rows of U=size of x
          no.of columns of U=size of t""

nr,nc=x.size,t.size
U=np.zeros([nr,nc])        #The necessary matrix
D=np.zeros([nr,nc])
for i in range (1,nr-1):
        U[i,0]=0           #Condition for t=0

for j in range(nc):
      U[nr-1,j]=u(t[j])
      
print(C*U[nr-1,nc-1])
c0=C*U[nr-1,nc-1]
a0=A*U[nr-1,0]
print(a0)
for j in range(0,nc-1):
      for i in range (0,nr-1):
         D[i,j]=(2*k*a+k*h*b)*U[i+1,j]+(4*h*h-4*k*a+2*k*c*h*h)*U[i,j]+(2*k*a-k*h*b)*U[i-1,j]
         if i==1:
             D[i,j]=D[i,1]-a0
         if i==nr-2:
             D[i,j]=D[i,nc-1]-c0
      print('d',D[:,j])
      W=np.linalg.solve(M,D[:,j])
      print('w',W)
      U[:,j+1]=W
for j in range(nc):
      U[0,j]=0
      U[nr-1,j]=u(t[j])
print(U)

T,X=np.meshgrid(t,x)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(T,X,U,color='red')
ax.set_xlabel("T")
ax.set_ylabel("X")
ax.set_zlabel("u")
ax.set_title("Solution of heat equation")
```
