```python
import scipy.integrate as integral
import numpy as np
import numpy.polynomial.laguerre as lag
import math
import matplotlib.pyplot as plt

#Defining Laguerre Function
def L(n,x):
    if n==0:
        return 1
    elif n==1:
        return 1-x
    else:
        return ((2*n-1-x)*L(n-1,x)-(n-1)*L(n-2,x))/(n)
    
#The problem:
#f(x)=(2+x^2)^(-1) 
#g(x)=exp(x)*(2+x^2)^(-1)

def f(x):
    f=1/(2+x**2)
    return f

def func(x):
    g=np.exp(x)*f(x)
    return g

dl_dx=lambda n,x:((x-n-1)*L(n,x)+(n+1)*L(n+1,x))/(x)


n=int(input("Enter your n:"))

#Finding roots of nth Laguerre Polynomial & storing it in t

Q=[]
for i in range (n):
    Q.append(0)

Q.append(1)
t=lag.lagroots(Q)

print("x (by quadrature rule)",t)


sol=0
for i in range(n):
    #weight
    w=-1/(n*dl_dx(n,t[i])*L(n-1,t[i]))
    sol1=w*func(t[i])
    sol=sol+sol1   
     
main_sol=sol
print("the value of integral of the function with n=",n,"is\n",main_sol)
man_sol=integral.quad(f, 0, math.inf)
man_sol=round(man_sol[0],10)
print("manual sol",man_sol)

y = lambda x:1/(2+x**2)
x = np.linspace(0,1000,1)
plt.plot(x,f(x))
```
